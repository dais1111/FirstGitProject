```python
#Iris分類問題
#特徴：雄しべと雌しべの花びらの長さと幅
#サンプル数：150
```

```python
import numpy as np
import scipy as sp
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
type(iris)
```

```python
iris.data
```

```python
x = iris.data[:, :2] #可視化できるよう二次元にする。列だけ絞る（０〜１）
y = iris.target #ラベル
print(x)
print(y)
```

```python
# データの標準化
mean = x.mean(axis=0) #平均
std = x.std(axis=0) #標準偏差
x = (x - mean)/std #正規化
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
clsf = SGDClassifier(max_iter=100) #確率的勾配下降法
clsf.fit(x, y) #学習？
```

```python
clsf.coef_
```

```python
clsf.intercept_
```

```python
# targetごとにファンシーインデックスで色分けして出す
plt.scatter(x[y==0,0],x[y==0,1],c="red")
plt.scatter(x[y==1,0],x[y==1,1],c="blue")
plt.scatter(x[y==2,0],x[y==2,1],c="green")

xx = np.linspace(-2.4, 3, 50)
plt.plot(xx, xx * clsf.coef_[0,0] + clsf.intercept_[0])
plt.plot(xx, xx * clsf.coef_[1,0] + clsf.intercept_[1])
plt.plot(xx, xx * clsf.coef_[2,0] + clsf.intercept_[2])
```

# 単回帰分析〜糖尿病患者予測〜

```python
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
diabetes_x = diabetes.data[:, np.newaxis, 2]
```

```python
diabetes.data
```

```python
diabetes_x
```

```python
x_train = diabetes_x[:-20]
y_train = diabetes.target[:-20]

x_test = diabetes_x[-20:]
y_test = diabetes.target[-20:]
```

```python
print(len(diabetes_x))
print(len(x_train))
print(len(x_test))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regr = LinearRegression()
regr.fit(x_train, y_train)
```

```python
y_pred = regr.predict(x_test)
y_pred
```

```python
#plt.scatter(x_train,y_train,c="green")
plt.scatter(x_test,y_test,c="red")

plt.plot(x_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```
